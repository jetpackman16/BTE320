```python
# Rock Paper Scissors Game
import random

play = input('Would you like to play a game (Yes/No)? ').capitalize()

while play == 'Yes':
  p1 = input('Player 1 - Choose between Rock/Paper/Scissors: ').capitalize()
  if p1 in ['Rock', 'Paper', 'Scissors']:
    p2 = random.choice(['Rock', 'Paper', 'Scissors']).capitalize()

    print(f'Player chose: {p1}; Computer chose: {p2} ')

    if p1 == p2:
      print("It's a tie!")
    elif ((p1 == 'Rock' and p2 == 'Scissors') or
        (p1 == 'Scissors' and p2 == 'Paper') or
        (p1 == 'Paper' and p2 == 'Rock')
        ):
      print('Player wins!')
    else:
      print('Computer wins!')

  else:
    print('Invalid Input')
  play = input('Would you like to play another game (Yes/No)? ').capitalize()
print('Have a good day!')
```

```python
def getValues(n):
    L = []
    for i in range(n):
        try:
            x = float(input("Enter a value: "))
            L.append(x)
        except ValueError:
            print("Invalid input. Please enter a number.")
            return []
    return L

def calcAverage(alist):
    if len(alist) == 0:
        print("No values were entered. Cannot calculate average.")
        return None
    sum = 0
    for x in alist:
        sum += x
    return sum / len(alist)

if __name__ == "__main__":
    try:
        n = int(input("Enter the number of values: "))
        L = getValues(n)
        avg = calcAverage(L)
        if avg is not None:
            print(f"The average is: {avg:.2f}")
    except ValueError:
        print("Invalid input. Please enter a whole number.")
```

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f'Hello, {self.first} {self.last}')
    else:
      print(f'Hi {self.first}')

  def __str__(self):
    #return str (self.first) + ", " + str(self.last)
    #return "{}, {}".format(self.first, self.last)
    return f'{self.first}, {self.last}'
greeting = Greeting('Donald', 'Becker')
greeting.display()
print(greeting)
```

```python
class Student:
    def __init__(self, name, number):
        self.name = name
        self._number = number
        self._courses = []

    def __len__(self):
        return len(self._courses)

    def enroll(self, course):
        if course in self._courses:
            print(f'You have already enrolled in {course}')
        else:
            self._courses.append(course)

    def get_courses(self):
        return self._courses

    def get_number(self):
        return self._number

    def set_number(self, newnumber):
        self._number = newnumber

s1 = Student('Bob', 100)
s2 = Student('Alice', 101)

print(s1.name, s2.name)
```

```python
class Coordinate:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    dist = (dx + dy) ** 0.5
    print(dist)

a = Coordinate(0,0)
b = Coordinate(3,4)
a.distance(b)
```

```python
# 2. a
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    return math.sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)
```

```python
#2. b
class Rectangle:
  def __init__(self):
    self.c1 = Point(0, 0)
    self.c2 = Point(0, 2)
    self.c3 = Point(2, 0)
    self.c4 = Point(2, 2)
    self.width = self.calculate_width()
    self.length = self.calculate_length()

  def calculate_width(self):
    return self.c1.distance(self.c2)

  def calculate_length(self):
    return self.c1.distance(self.c3)

  def area(self):
    return f'Area = {self.width * self.length} sq. ft.'

square = Rectangle()
square.area()
```
